## Python Libraries, Global Variables, and Functions

```python
## Global Variables and all the libraries used in this lab session 

import math

global lamHeNe # wavelength of Helium Neon laser in metres

lamHeNe = 6.328e-7 #m

# important data that is collected from the setup 
global di
global ddi
global do
global ddo

global di_mc
global ddi_mc
global do_mc
global ddo_mc

# Calculated information
global f  # focal length
global df # uncertainty of focal length

global m  # magnification
global dm # uncertainty of magnification

global mc # magnification of converging len
global dmc # uncertainty of converging magnification

global M # total magnification 
global dM # uncertainty of total magnification

global d  # separation between two peaks 
global dd # uncertainty of separation between two peaks 

def calculateFocal():
    global f
    global df

    f = di*do/(do + di)
    df_ddi = (do/(do+di))**2
    df_ddo = (di/(do+di))**2
    df = math.sqrt( (df_ddi*ddi)**2 + (df_ddo*ddo)**2 )
    display(HTML("Experimental focal length is %0.02f  %0.02f cm" % (f,df)))

def calculateMag():
    global m
    global dm

    m = di/do
    dm_ddi = 1/do
    dm_ddo = -di/(do**2) 
    dm =math.sqrt( (dm_ddi*ddi)**2 + (dm_ddo*ddo)**2 )
    display(HTML("Experimental magnification is %0.03f  %0.03f" % (m,dm)))

def calculateMagC():
    global mc
    global dmc

    mc = di_mc/do_mc
    dm_ddi = 1/do_mc
    dm_ddo = -di_mc/(do_mc**2) 
    dmc = math.sqrt( (dm_ddi*ddi)**2 + (dm_ddo*ddo)**2 )
    display(HTML("Experimental converging magnification is %0.03f  %0.03f" % (mc,dmc)))

def calculateFocalC():
    f = di_mc*do_mc/(do_mc + di_mc)
    df_ddi = (do_mc/(do_mc+di_mc))**2
    df_ddo = (di_mc/(do_mc+di_mc))**2
    df = math.sqrt( (df_ddi*ddi)**2 + (df_ddo*ddo)**2 )
    display(HTML("Experimental converging focal length is %0.02f  %0.02f cm" % (f,df)))

def calculateTotalM():
    global mc
    global dmc
    global m
    global dm
    global M
    global dM

    M = m*mc
    dM_dm = mc
    dM_dmc = m
    dM = math.sqrt( (dM_dm*dm)**2 + (dM_dmc*dmc)**2)
    display(HTML("Experimental total magnification of the mesh system is %0.02f  %0.02f" % (M,dM)))

def calculateSpatialFreq():

    global M
    global dM
    
    global f
    global df
    
    global d
    global dd

    global lamHeNe
    

    f_m = f/100     # focal length in metre
    df_m = df/100 # focal length in metre

    nu = d/(M*f_m*lamHeNe) # 1/m

    dnu_dd = 1/(M*f_m*lamHeNe)
    dnu_dM = -d/((M**2) * f_m * lamHeNe)
    dnu_df = -d/(M * (f_m**2) * lamHeNe)

    dnu = math.sqrt( (dnu_dd * dd)**2 + (dnu_dM * dM)**2 + (dnu_df * df_m)**2 )
    display(HTML("The calculated spatial frequency is %0.03f  %0.03f $m^{-1}$ " %(nu, dnu)))
    wire_spacing = 1/nu  #m/wire
    dwire_spacing = (1/(nu**2))*dnu
    display(HTML("The calculated wire spacing is %0.03f  %0.03f  mm " %(wire_spacing*1000, dwire_spacing*1000)))
    actual_wire_spacing = (1/40)* (1/100) # m /wire
    display(HTML("The wire spacing according to speculation %0.02f mm " % (actual_wire_spacing*1000)))
    error = 100 * abs(actual_wire_spacing -wire_spacing)/actual_wire_spacing
    display(HTML("Error between actual and calculated is %0.02f %%" %(error)))
```

Using the optical imaging system seen in the image above, we used the following 1st year physics equations to determine the overall magnification of the system, and the focal length of its lens.  

#### Error Anaylsis of magnification ($\bf{m}$)

$m = \frac{d_{i}}{d_{o}}$

$\frac{ \partial m}{\partial d_{i}} =  \frac{1}{d_{o}}$


$\frac{\partial m}{\partial d_{o}} =  -\frac{d_{i}}{ d_{o}^2}$



$\delta m = \sqrt{ (\frac{\partial m}{\partial d_{i}} * (\delta d_{i}))^2  + (\frac{\partial m}{\partial d_{o}} * (\delta d_{o}))^2 }$



#### Error Anaylsis of focal length ($\bf{f}$)

$f = \frac{d_{i}d_{o}}{d_{i} + d_{o}} $

$\frac{\partial f}{\partial d_{i}} =  \frac{d_{o}^2}{ (d_{i} + d_{o})^2}$


$\frac{\partial f}{\partial d_{o}} =  \frac{d_{i}^2}{ (d_{i} + d_{o})^2}$


$\delta f = \sqrt{ (\frac{\partial f}{\partial s_{i}} * (\delta s_{i}))^2  + (\frac{\partial f}{\partial s_{o}} * (\delta s_{o}))^2 }$

### Verifying the Mesh's Spatial Frequency ($\bf{\nu}$)

The separation between the bright spots (in both x and y positions) is measured to be $d_{x} = d_{y} = 2.2 \pm 0.2 cm $.

Like I said earlier, the converging len converts the positions of the incoming wave $(x,y)$ to $(\theta_{x} f, \theta_{y} f)$ where f is the focal lengths.

$\theta_{x}$ and $\theta_{y}$ are the angles that the wavevector $\bf{k}$ makes with x-z and y-z planes respectively. To be more precise, $\theta_{x} = \arcsin{\frac{k_{x}}{\bf{k}}}$ and $\theta_{y} = \arcsin{\frac{k_{y}}{\bf{k}}}$ . We can further simpliy these two terms. The incoming beam is a collimated Gaussian beam, so we can assume that it's paraxial, so $\theta_{x} \approx \frac{k_{x}}{\bf{k}} $ and $\theta_{y} \approx \frac{k_{y}}{\bf{k}}$ . Furthmore, since there's an equal spacing in both x and y direction, we can say that $\theta = \theta_{x} = \theta_{y}$ and consequently, $\nu = \nu_{x} = \nu_{y}$ and $k = k_{x} = k_{y}$


To find the spatial frequency, we can rewrite $\theta$ in terms of $\nu$. 

$$ \theta = \frac{k}{\bf{k}} = \frac{k\lambda}{2\pi} =  \nu \lambda $$ 


We need to remember that our measurement of $d_{x} = d_{y} = d = \theta f$ is the magnified measurement of the separation between the bright spots. So we need to divide $d$ by the total magnification ($M$).

The net magnification  is a product of two magnification constants that I calculated above. 


$$\theta = \nu \lambda = \frac{d}{Mf} $$



$$\nu = \frac{d}{Mf\lambda} $$

Then the wire spacing (W) is : 

$$W = \frac{1}{\nu}$$



#### Error Anaylsis of Spatial Frequency


$ \frac{\partial \nu }{\partial d} =\frac{1}{Mf\lambda} $

$ \frac{\partial \nu }{\partial M } = -\frac{d}{M^2 f\lambda} $

$ \frac{\partial \nu }{\partial f } = -\frac{d}{M f^2 \lambda} $


$$\delta \nu = \sqrt{ (\frac{\partial \nu}{\partial d} * (\delta d))^2  + (\frac{\partial \nu}{\partial M} * (\delta M))^2 + (\frac{\partial \nu}{\partial f} * (\delta f))^2 }$$


#### Error Anaylsis of Wire Spacing ($\bf{W}$) From Spatial Frequency

$ \frac{\partial W}{\partial \nu} = -\frac{1}{\nu^2}$

$$ \delta W = \mid \frac{\partial W}{\partial \nu}\mid * (\delta \nu) $$

Since we want x axis high frequency light, we want a horizontal slit, and since we want y = x axis high frequency light we want another slit at approximately 45 degrees.

## Dark Field Image

According to this link https://www.microscopyu.com/techniques/stereomicroscopy/darkfield-illumination, dark field image takes out the zeroth order (or the brightest/central spot) from the differaction pattern. What this means is a complete removal of any constant light in an image.  An example would be an image of a chair on a white background, when the zeroth order spatial frequencies are removed all that remains are the higher spatial frequenies, completely darkening (and removing) the background of the image.  All that would remain is light along the outline of the chair.  

This will remove zeroth order spatial frequency of light. This implies that only spatial frequencies above a threshold can pass through. In order to produce a dark field image, we use a transparent plane with a black (absorbing dot) in the very middle, that is placed exactly at the fourier transform plane where all the spatial frequencies are spread out.  


According to both textbooks(*Classical and Modern Optics*  and *Fundamental of Photonics*), the high pass filter blocks low frequency and transmits high frequency. As a result, the edge in the image gets enhanced. From both textbooks, a black dot filter will cause this dark field image.

### Phase Contrast

#### Background:  

A phase plate is a transparent plate with varying thickness in accordance with a square wave.  The thickness of it varies along the x axis, and as a result of it being some medium that is not air, causes a phase shift due to different distances travelled through it.  This phase shift in the E field is normally not visible, since the magnitude of E doesn't change appreciably.  However there is a method in order to be able to visualize slight changes to phase shifts in E.  This is extremely useful when imaging transparent materials that cause suble phase shifts due to varying thicknesses.  

The method used to show phase shifts involves shifting the zeroth spatial frequency, by the placement of a thin film of some thickness at the very beam center.  This causes the constant E field to shift closer to the phase shift caused by the object.  Converting this to intensity by squaring it produces an intensity that is far more sensitive to phase shifts than before.

The specifications of the phase grating say a distance of 0.3 mm per half period.  We have measured the following:

$$ \frac{7 Periods}{3.6mm}$$

We can then say the following:

$$ \frac{0.3mm \times 2}{Period} \times m = \frac{3.6mm}{7 periods}$$

Solving for magnification m gives us:

$$m = \frac{5.142 mm / Period}{0.6 mm / Period} = 8.57$$

This compared favourably with our previous measurement for the magnification being:  $m = 8 \pm 5$
