## Purpose
The purpose of this lab session is to investigate the difference between Fraunhofer and Fresnel Diffraction regimes.  For this experiment we shall use a variable width slit aperture, with collimated light being shone through it, to observe the diffraction patterns at some distance from the aperture.  

Based on the width of the aperture, we shall see diffraction patterns corresponsing to Fresnel and Fraunhofer regimes.

## Background/Theory

Fraunhofer diffraction occurs with the paraxial approximation, and the approximation that the object being imaged is much smaller than the resulting diffraction pattern.  

Fresnel diffraction occurs again with the paraxial approximation, but doesn't require that the object being images is much smaller than the resulting diffraction pattern.

The variable which determines which regime the diffraction lies within is:

$$ \Delta v = w(\frac{2}{R \lambda})^\frac{1}{2} $$
$$ w $$ is the slit width
$$ R $$ is the distance from slit to diffraction pattern
$$ \lambda $$ is the wavelength of light

When $$ \Delta v $$ is less than 1, we can approximate the E field oscillations as plane waves, however when it is greater than 1 we must consider the oscillations to be in radial directions.  This leads to the different diffraction regimes.  

### Fraunhofer Model:

In order to model the diffraction pattern we start with the imposed E field configuration:

$$ E^{(+)} = A  $$ if |x| is less than or equal to half the slit width, and zero otherwise.

Next we calculate the fourier transform, only in the x axis:

$$ E'^{(+)} = \int_{-\infty}^\infty \! E^{(+)} e^{-i k_x x} \, \mathrm{d}x = A \int_{-w/2}^{w/2} \! e^{-i k_x x} \, \mathrm{d}x $$

Doing the integration we arrive at an expression for E:

$$ E^{(+)}(x,z) = \sqrt{\frac{2z}{\pi k}}\frac{1}{x} A sin(\frac{kxw}{2z})$$

Squaring this number, we will arrive at an expression that is proportional to the real intensity.

#### Analysis:

In order to compare these diffraction patterns to the theoretical diffraction patterns described earlier, we need to extract the intensity profile of the pattern from the right to the left hand side of each image.  We will use the MATLAB image processing toolbox to accomplish this task.  

**One potential issue here is that the diffraction patterns in the images are not completely normal to the axis we shall be looking across.  This may affect the intensity profiles extracted depending on the method used!**

##### Intensity Extraction Method:

In order to extract the intensity we shall use the following procedure using MATLAB:

1.  Convert the image to greyscale
2.  Adjust the intensity range to increase the usable data (imadjust)
3.  Rotate the image such that the diffraction pattern is normal to the axis we're measuring intensity along
4.  Average the intensity along the y axis (some portion of the y axis)
5.  Plot the intensity profile

It should be noted that the imadjust will probably change the magnitude of the intensity being viewed.  In addition, all spatial scales are no longer valid given that the images have been cropped.  We shall simply fit the data as best we can, without regard to the correct scale.  

The following is the matlab code used to generate the theoretical model numbers and process the intensity profiles from the experiment.  

```python
% Generating theoretical Model for Fresnel
clear all;
close all;

lambda = 6.328e-7;
k = (2*pi())/lambda;
z = 3.93;

j = 1;
scales = [3 3 3 1 0.5 0.5];

figure;
for w = [0.2 0.5 0.91 1.15 4.77 6.01]./1000 %m
    %% Processing Images
    
    [x,inten] = intensity_profile(j);
    
    subplot(2,3,j)
    plot(x,inten,'g');
    hold on;
    xlim([-1 * scales(j),1 * scales(j)]);
    hold on;
    
    %% Generating Theoretical Models
    x = linspace(-0.01 * scales(j),0.01 * scales(j),1000);
    
    integrals = c_int( sqrt(k./(pi().*z)) .* (x + (w/2))) - c_int( sqrt(k./(pi().*z)) .* (x - (w/2))) + i .* s_int( sqrt(k./(pi().*z)) .* (x + (w/2))) - i .* s_int( sqrt(k./(pi().*z)) .* (x - (w/2)));
    
    E_fresn = abs(1 .* sqrt(-i/2) .* exp(i .* k .* z) .* integrals);
    E_fraun = sqrt(2*z * (pi() * k)^-1) .* x.^-1 .* sin( w .* k .* x .* (2 * z)^-1);
    
    Es_fresn = E_fresn .^ 2;
    Es_fraun = E_fraun .^ 2;
    
    
    plot(x.*100,Es_fresn,'-r','linewidth',3);
    hold on;
    plot(x.*100,Es_fraun,'b','linewidth',2);
    hold on;
    xlabel('X position on Screen (cm)');
    ylabel('E^2 (proportional to Intensity)');
    title(strcat('Theoretical Diffraction Intensity, Slit Width: ',num2str(w.*1000),'mm'));
    legend('Experimental Result','Fresnel Model','Fraunhofer Model');
    hold off;
    j = j+1;
end

function [x,inten] = intensity_profile(test_num)
    
    % Fitting Parameters
    vsf = [0.01 0.015 0.03 0.55 1.8 1.8];
    hsf = [0.35 0.25 0.2 0.2 0.15 0.2];
    hshift = [0 0 -0.07 0 0 0];
    
    name=strcat('snap_',num2str(test_num),'.jpeg');
    
    % Reading Image
    I = imread(name);
    
    % Converting to Greyscale
    I_grey = imadjust(rgb2gray(I));
    
    inten = zeros(size(I_grey,2),1);
    j = 1:size(inten,1);
    middle = round(size(I_grey,1)./2,0);
    dE = 5;
    %inten(j) = sum(I_grey(:,j)) ./ (size(I_grey,1));
    inten(j) = sum(I_grey(middle-dE:middle+dE,j)) ./ 11;
    x = linspace(-10.*hsf(test_num),10.*hsf(test_num),size(inten,1));
    x = x + hshift(test_num);
    inten = inten .* vsf(test_num).* 255^-1;
end
    
function c = c_int(x)
    c = zeros(1,size(x,2));
    for i = 1:size(x,2)
        fun = @(t) cos( (pi()*t.^2)./2 );
        c(i) = integral(fun,0,x(i));
    end

end

function s = s_int(x)
    s = zeros(1,size(x,2));
    for i = 1:size(x,2)
        fun = @(t) sin( (pi()*t.^2)./2 );
        s(i) = integral(fun,0,x(i));
    end

end
```
